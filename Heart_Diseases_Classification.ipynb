```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Visualize the dataset before PCA
sns.pairplot(scaled_df)
plt.show()

# 2. Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)

# Create a DataFrame for the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])


# 3. Visualize the transformed data
plt.figure(figsize=(8, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Heart Disease Dataset')
plt.show()

# 4. Plot the principal component axes (eigenvectors)
# Get the eigenvectors (principal component axes)
eigenvectors = pca.components_

# Calculate the mean of the original data
data_mean = scaled_df.mean()

# Scale the eigenvectors manually
scale_factor = 3.5 # Adjust this value to control the length of the eigenvectors
scaled_eigenvectors = eigenvectors * scale_factor

# Plot the eigenvectors as lines starting from the data mean
plt.figure(figsize=(8, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], label='Original Components', alpha=0.6) # Plot the data points
plt.quiver(data_mean[0], data_mean[1], scaled_eigenvectors[0, 0], scaled_eigenvectors[0, 1], angles='xy', scale_units='xy', scale=1, color='r', label='PC1')
plt.quiver(data_mean[0], data_mean[1], scaled_eigenvectors[1, 0], scaled_eigenvectors[1, 1], angles='xy', scale_units='xy', scale=1, color='b', label='PC2')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA with Eigenvectors')
plt.legend()
plt.show()


# EXTRA
# Finally, We will check the orthogonality of the eigenvectors
# Normalize the eigenvectors to unit vectors
unit_eigenvectors = eigenvectors / np.linalg.norm(eigenvectors, axis=1)[:, np.newaxis]


# Check if the unit eigenvectors are orthogonal
dot_product = np.dot(unit_eigenvectors[0], unit_eigenvectors[1])

# Print the result
print("Unit Eigenvectors:")
print(unit_eigenvectors)
print("\nDot Product of Unit Eigenvectors:", dot_product)

if np.isclose(dot_product, 0):
    print("The eigenvectors are orthogonal.")
else:
    print("The eigenvectors are not orthogonal.")
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Identify all categorical columns (optional if known)
categorical_cols = df.select_dtypes(include=['object']).columns

# Define nominal and ordinal columns
nominal_cols = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina']
ordinal_cols = ['ST_Slope']

#testing
# Print unique values of each categorical column
print("Categorical Columns and their unique values:")
for col in categorical_cols:
    print(f"Column: {col}")
    print(f"Unique values: {df[col].unique()}")
    print("-" * 30)

# 1. OneHotEncode nominal columns
df_nominal = pd.get_dummies(df[nominal_cols], drop_first=False)

# 2. LabelEncode ordinal columns
for col in ordinal_cols:
    df[col] = label_encoder.fit_transform(df[col])

# 3. Drop original nominal columns
df.drop(columns=nominal_cols, inplace=True)

# 4. Concatenate the one-hot encoded nominal columns back into df
df = pd.concat([df, df_nominal], axis=1)

# Print final DataFrame
print(df.head())
```

```python
#(b)
from sklearn.model_selection import train_test_split

# Assuming df1 is our DataFrame with label encoded categorical features
x = df[['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'ST_Slope', # Numerical Features
       'Sex_F', 'Sex_M', 'ChestPainType_ASY', 'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA', # One-hot encoded nominal features
       'RestingECG_LVH', 'RestingECG_Normal', 'RestingECG_ST',
       'ExerciseAngina_N', 'ExerciseAngina_Y']]
y = df['HeartDisease']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=48) # 80% training and 20% test

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

import pandas as pd

# Check imbalance level (To know to use SMOTE or not)
"""
class_ratio = y_train.value_counts()[0] / y_train.value_counts()[1]
print(f"Minority Ratio: {class_ratio:.2f}")

print("Class counts in training set:", y_train.value_counts())

Conc: healthy class (410 samples) is 80.7% the size of the heart disease class (508),
  Thus, dataset is not imbalanced.
  Therfore, we don't need to use SMOTE.
"""

```
